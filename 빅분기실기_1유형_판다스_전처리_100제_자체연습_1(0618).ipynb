```python
import pandas as pd
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)

# 21. quantity 컬럼 값이 3인 데이터 5행 출력
cond = df['quantity'] == 3
print(df[cond].head(5))

# 22. 컬럼 값이 3인 데이터 추출, index 정렬하고 첫 5행 출력
print(df[cond].reset_index(drop = True).head(5))

# 23. quantity, item_price로 구성된 새로운 데이터 프레임 정의
new_data = pd.concat([df['quantity'], df['item_price']], axis = 1)
print(new_data)

new_data2 = df[['quantity', 'item_price']]
print(new_data2)

# 24. item_price 컬럼의 달러표시 문자 제거. float타입으로 저장하여 new_price에 대입
df['new_price'] = df['item_price'].str.replace('$', '').astype('float')
print(df['new_price'])

# 25. new_price 컬럼이 5 이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수 구하기
cond = df['new_price'] <= 5
print(df[cond])
print('전체 갯수:',  len(df[cond]))
```

```python
# 26. item_name 명이 Chicken Salad Bowl인 데이터 프레임을 추출하고, index값 초기화.
cond1 = df['item_name'] == 'Chicken Salad Bowl'
#print(df[cond1].reset_index(drop=True).head())

# 27. new_price 값이 9 이하이고, item_name 값이 Chicken Salad Bowl인 데이터 프레임 추출
cond2 = df['new_price'] <= 9
cond3 = df['item_name'] == 'Chicken Salad Bowl'

#print(df[cond2 & cond3].head())

# 28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화.
print(df['new_price'].sort_values(ascending = True).reset_index(drop=True))

# 29. df의 item_name 컬럼 값 중 Chips를 포함하는 경우의 데이터 출력
cond4 = df['item_name'].str.contains('Chips')
print(df[cond4])
#contains 자주 까먹는데, 잊어버리지 말것!

# 30. df의 짝수번째 컬럼만을 포함하는 데이터프레임 출력
print(df.iloc[:,::2].head())
```

```python
# 31. df의 new_price 값에 따라 내림차순 정렬 / index 초기화
print(df['new_price'].sort_values(ascending = False).reset_index(drop=True))

# 32. df의 item_name 칼럼이 Steak Salad 또는 Bowl인 데이터 인덱싱

cond1 = df['item_name'] == 'Steak Salad'
cond2 = df['item_name'] == 'Bowl'
print(df[cond1 | cond2].reset_index(drop=True))

# 33. 위 데이터를 데이터프레임화하고, item_name을 기준으로 중복행이 있으면 제거
new_data = df[cond1 | cond2]
print(new_data.drop_duplicates('item_name'))
# drop_duplicates로 중복행 제거가 가능.

# 34. 위 데이터 중 마지막 케이스만 남기기
print((df.drop_duplicates('item_name', keep = 'last')))

# 35. df의 데이터 중 new_price값이 평균값 이상
cond3 = df['new_price'] >= df['new_price'].mean()
print(df['new_price'].mean())
print(df[cond3])
```

```python
# 36. df 데이터 중 item_name의 값이 lzze 데이터를 Fizzy Lizzy로 수정

df['item_name'] = df['item_name'].str.replace('Izze', 'Fizzy Lizzy')
print(df['item_name'].head(20))
#str.replace 사용시 대입 주의!

# 37. df 데이터 중 choice_description 값이 Nan인 데이터의 갯수
print(df['choice_description'].isnull().sum())

# 38. df의 데이터 중 choice_description 값을 NoData 값으로 대체
df['choice_description'] = df['choice_description'].fillna('NoData')
print(df['choice_description'])

# 39. df의 데이터 중 choice_description 값에 Black이 들어가는 경우 인덱싱
cond1 = df['choice_description'].str.contains('Black')
print(df[cond1])

# 40. df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수
cond2 = df['choice_description'].str.contains('Vegetables')
print(df[~cond2])
```

```python
# 46. 데이터의 각 host_name의 빈도수 출력 / 빈도수 기준 내림차순 정렬 *** 초어려움!!!!
dfGroup2 = df.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending = False)
print(dfGroup2)

# 47. neighbourhood_group의 값에 따른 neighbourhood 컬럼 값의 갯수를 구하여라
print(df.groupby(['neighbourhood_group', 'neighbourhood']).size())

# 48. neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group 그룹의 최댓값
print(df.groupby(['neighbourhood_group', 'neighbourhood']).size().groupby(['neighbourhood_group']).max())

# 49. neighbourhood_group값에 따른 price값의 평균, 분산, 최대, 최소값 구하기
print(df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').describe())

print(df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').describe())
```
