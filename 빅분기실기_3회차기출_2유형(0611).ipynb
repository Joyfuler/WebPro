## 데이터 불러오기

```python
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

## EDA

```python
# 크기 확인
print(train.shape, test.shape)
#test의 컬럼이 하나 적다는 걸 알 수 있음.
print(train.head())
print('--------------------------')
print(test.head())
# 따라서 target 값은 train의 travelinsurance
```

```python
# 타입 확인
print(train.info()) # 오브젝트형 4개, test와 train 동일.
print('----------')
print(test.info())
```

```python
# 결측치 확인
print(train.isnull().sum())
print(test.isnull().sum())

# train / test 모두 결측치는 없음을 알 수 있음.
```

```python
# 기초 통계 획인
print(train.describe())
print('-------------------')
print(test.describe())
# outliar는 딱히 없는 것으로 추정. 연령대는 20대~ 30대로 분류하기가 애매
```

```python
# 기초 통계 획인 object
print(test.describe(include = 'object'))
print(train.describe(include = 'object'))

# object값은 기본적으로 0/1 두가지 경우의 수로 이뤄져 있음
```

```python
# 타겟(레이블)
print(train['TravelInsurance'].value_counts())
# 비가입 965 / 가입 525
```

## 데이터 전처리 및 피처엔지니어링

```python
# 수치형 데이터와 범주형 데이터 분리
n_train = train.select_dtypes(exclude = 'object').copy()
c_train = train.select_dtypes(include = 'object').copy()
n_test = test.select_dtypes(exclude = 'object').copy()
c_test = test.select_dtypes(include = 'object').copy()

print(n_train.shape, c_train.shape, n_test.shape, n_train.shape)
```

```python
# 수치형 변수 스케일링
from sklearn.preprocessing import MinMaxScaler
# 수치형 컬럼을 우선 가져온다
cols = ['Age', 'AnnualIncome', 'FamilyMembers', 'ChronicDiseases']
print(cols)
scaler = MinMaxScaler()

n_train[cols] = scaler.fit_transform(n_train[cols])
n_test[cols] = scaler.transform(n_test[cols])
print(n_train.head())
print(n_test.head())
```

```python
# 범주형 변수 인코딩 - 원핫인코딩 사용 (get_dummies)
display(c_train.head())
c_train = pd.get_dummies(c_train)
c_test = pd.get_dummies(c_test)
print(c_train.head())
```

```python
# 분리한 데이터 다시 합침
train = pd.concat([n_train, c_train], axis = 1)
test = pd.concat([n_test, c_test], axis = 1)
print(train.head())
print(test.head())
```

## 검증데이터 분리

```python
# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train.drop(['TravelInsurance'], axis = 1),
                                            train['TravelInsurance'],
                                            test_size = 0.1,
                                            random_state = 2024) #검증 데이터 + 타겟값
print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)

# 검증 데이터는 1341 / 테스트 데이터는 149개.
```

## 모델 학습 및 예측

```python
# 랜덤포레스트 (분류문제. 보험 가입인지 아닌지를 묻는 문제이므로)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 400, max_depth = 9, random_state = 2024)
model.fit(X_tr, y_tr)
pred = model.predict_proba(X_val)
print(pred[:,1])

# roc_auc_score를 가져와 평가. 0.7487329434697856
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_val, pred[:,1]))
```

## 예측
